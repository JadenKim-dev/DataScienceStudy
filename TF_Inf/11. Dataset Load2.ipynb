 ## 2. tfds.load() : dataset의 형식

```python
dataset = tfds.load(name='mnist', shuffle_files=True)

print(type(dataset))
```

```python
print(dataset.keys())
print(dataset.values())
```

```python
train_ds = dataset['train']
test_ds = dataset['test']

print(type(train_ds))
print(type(test_ds))
```

```python
for tmp in train_ds:
    print(type(tmp))
    print(tmp.keys())

    break
```

```python
for tmp in train_ds:
    images = tmp['image']
    labels = tmp['label']

    print(images.shape)
    print(labels.shape)

    break
```

```python
# dataset에 batch를 지정할 경우, batch를 단위로 데이터를 뽑아온다

train_ds = dataset['train'].batch(32)
for tmp in train_ds:
    images = tmp['image']
    labels = tmp['label']

    print(images.shape)
    print(labels.shape)

    break
```

## 3. tfds.load() : as_supervised=True 지정

```python
dataset = tfds.load(name='mnist', 
                    shuffle_files=True,
                    as_supervised=True) 
train_ds = dataset['train'].batch(32)
test_ds = dataset['test']
```

```python
for tmp in train_ds:
    print(type(tmp))   # as_supervised=True 지정시 tuple이 들어있음

    break
```

```python
for tmp in train_ds:
    images = tmp[0]
    labels = tmp[1]

    print(images.shape)
    print(labels.shape)

    break
```

```python
# as_supervised=True 지정 시, 사용 가능한 표현
for images, labels in train_ds:
    print(images.shape)
    print(labels.shape)

    break
```

## 4. tfds.load(): split 지정

```python
train_ds, test_ds = tfds.load(name='mnist', 
                    shuffle_files=True,
                    as_supervised=True,
                    split=['train', 'test']) 

train_ds = train_ds.batch(32)
for images, labels in train_ds:
    print(images.shape)
    print(labels.shape)
    break
```

### with_info=True, split 함께 지정시

```python
# train_ds와 test_ds는 튜플로 묶어줘야 함

(train_ds, test_ds), ds_info = tfds.load(name='mnist', 
                    shuffle_files=True,
                    as_supervised=True,
                    split=['train', 'test'],
                    with_info=True) 
print(train_ds)
print(test_ds)
print(ds_info)
```
