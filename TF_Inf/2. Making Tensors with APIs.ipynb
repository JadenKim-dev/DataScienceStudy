## 2. tf.ones_like(), tf.zeros_like()

```python
test_list = [[1, 2, 3], [4, 5, 6]]
t1 = tf.Variable(test_list)
print(t1)
```

```python
t2 = tf.ones(shape=t1.shape)
print(t2)
```

```python
t2 = tf.ones_like(t1)
print(t2)
```

```python
t3 = tf.zeros_like(t1)
print(t3)
```

## 3. tf.random 

```python
# tf.random.set_seed(0)  # -> 값 고정 가능
t1 = tf.random.normal(shape=(10, 10))
print(t1)
```

```python
import matplotlib.pyplot as plt
t2 = tf.random.normal(mean=5, stddev=3, shape=(1000, ))
print(t2[:100])

fig,ax = plt.subplots(figsize=(5, 5))
ax.hist(t2.numpy(), bins=30)

ax.tick_params(labelsize=15)
```

```python
t3 = tf.random.uniform(shape=(1000,), minval=-10, maxval=10)

fig,ax = plt.subplots(figsize=(5, 5))
ax.hist(t2.numpy(), bins=30)

ax.tick_params(labelsize=15)
```

## 4. Tensor.dtype, Tensor.shape

```python
t1 = tf.random.normal(shape=(128, 128, 3))

print("t1.shape: ", t1.shape)
print("t1.dtype: ", t1.dtype)
```

```python
test_np = np.random.normal(0, 1, size=(100, ))
print(test_np.dtype)
```

```python
test_np = np.random.randint(-10, 10, size=(100,))
print(test_np.dtype)
```

```python
t1 = tf.constant(test_np)
print(t1.dtype)
```

=> int64인 상태에서 Tensor로 변환하니, 해당 Tensor의 dtype도 그대로 int64가 된다  
=> 이럴 경우 학습이 제대로 이루어지지 못할 수 있다

```python
test_np = np.random.randint(-10, 10, size=(100,))
t1 = tf.constant(test_np, dtype=tf.float32)
print(t1.dtype)
```
