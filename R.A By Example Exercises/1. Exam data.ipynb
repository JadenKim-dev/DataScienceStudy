```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
```

## Loading Data

```python
exam = pd.read_csv("./data/1-examination.csv")
exam.head()
```

## Fitting Model 1 (F = B0 + B1P1)

```python
target = exam[['F']]
target.head()
```

```python
x_data = exam[['P1']]
x_data.head()
```

```python
x_data1 = sm.add_constant(x_data, has_constant="add")
model1 = sm.OLS(target, x_data1)
fitted_model1 = model1.fit()
fitted_model1.summary()
```

## Fitting Model 2 (F = B0 + B2P2)

```python
x_data2 = exam[['P2']]
x_data2.head()
```

```python
x_data2 = sm.add_constant(x_data2, has_constant="add")
model2 = sm.OLS(target, x_data2)
fitted_model2 = model2.fit()
fitted_model2.summary()
```

## Fitting Model 3 (F = B0 + B1P1 + B2P2)

```python
x_data3 = exam[['P1', 'P2']]
x_data3.head()
```

```python
x_data3 = sm.add_constant(x_data3, has_constant="add")
model3 = sm.OLS(target, x_data3)
fitted_model3 = model3.fit()
fitted_model3.summary()
```
