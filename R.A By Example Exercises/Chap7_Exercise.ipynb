```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
from statsmodels.stats.anova import anova_lm
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("./data/7_Education.csv")
df.head()
```

```python
df = df.drop(48).reset_index(drop=True)
```

```python
df.tail()
```

```python
edu = pd.get_dummies(df, columns=['Region'], drop_first=True)
edu.head()
```

```python
target = edu['Y']
x_data1 = edu.iloc[:, 2:]
x_data1_1 = sm.add_constant(x_data1, has_constant="add")
model1 = sm.OLS(target, x_data1_1)
fitted_model1 = model1.fit()
fitted_model1.summary()
```

```python
target = edu['Y']
x_data2 = edu.iloc[:, 2:5]
x_data2_1 = sm.add_constant(x_data2, has_constant="add")
model2 = sm.OLS(target, x_data2_1)
fitted_model2 = model2.fit()
fitted_model2.summary()
```

```python
fitted1 = fitted_model1.predict(x_data1_1)
fm_sse = 0
target_val = edu['Y']
for i in range(target_val.shape[0]):
    fm_sse += (fitted1[i] - target_val[i])**2
fm_sse
```

```python
fitted2 = fitted_model2.predict(x_data2_1)
rm_sse = 0
target_val = edu['Y']
for i in range(target_val.shape[0]):
    rm_sse += (fitted2[i] - target_val[i])**2
rm_sse
```

```python
from scipy.stats import f
F_stats = ((rm_sse-fm_sse)/3)/(fm_sse/(49-7))
F_stats
```

```python
f_dist = f(3, 49-7)
1-f_dist.cdf(F_stats)
```
